```python
df = pd.read_csv('swipe_only.csv')
df.head()
```

```python
swipes = {}
for i in np.arange(df.shape[1]):
  swipes.update({df.columns[i]: list(df.iloc[:, i].value_counts())})
```

```python
type(swipes['swipe10_c_f_a'])
```

```python
for key in swipes.keys():
  if len(swipes[key]) == 1:
    swipes[key].append(0)
swipes
```

```python
swipe_df = pd.DataFrame(data = swipes)
```

```python
swipe_df2 = swipe_df.transpose().rename(columns = {0: 'winning_count', 1: 'losing_count'})
```

```python
swipe_df2.to_csv('swipe_counts.csv')
```
